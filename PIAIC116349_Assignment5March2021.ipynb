# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from os import listdir
```

# Uploading Data

```python
li=[]
data_state=[]
for f in listdir("./"):
    if f.endswith(".csv"):
        if f!="inventory.csv":
            li.append(f)
print(li)
for i in li:
    df=pd.read_csv(i,index_col=None,header=0)
    data_state.append(df)
    
    
 
```

* This will automatically upload all the data 

# Concatenation of all the data row wise

```python
us_census=pd.concat(data_state,axis=0,ignore_index=True)
us_census
```
